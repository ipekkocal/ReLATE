```python
#Spider Chart- Case 1

import matplotlib.pyplot as plt
import numpy as np

cases = [
    "Case 1"
]
min_values = [32.51]
max_values = [91.78]
relate_values = [88.49]

categories = cases * 3  
min_values = min_values * 3
max_values = max_values * 3
relate_values = relate_values * 3

data = np.array([min_values, max_values, relate_values])

num_vars = len(categories)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  

data = np.concatenate((data, data[:, :1]), axis=1)

plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)

labels = ["Min", "Max", "ReLATE"]
colors = ["red", "green", "blue"]
for i in range(data.shape[0]):
    ax.fill(angles, data[i], alpha=0.25, color=colors[i])
    ax.plot(angles, data[i], label=labels[i], color=colors[i])

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=16)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(["20", "40", "60", "80", "100"], fontsize=12)
ax.set_ylim(0, 100)

plt.legend(loc="lower right", bbox_to_anchor=(1.3, 0.05), fontsize=16)  # Adjust bbox_to_anchor as needed
plt.title("Spider Chart for Min, Max, and ReLATE Accuracy Percentages", fontsize=24, fontweight="bold")
plt.tight_layout()

plt.show()
```
